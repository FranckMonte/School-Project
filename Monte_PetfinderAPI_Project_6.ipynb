```python
# What type of animals are available for adoption in Virginia?
url_type = 'https://api.petfinder.com/v2/types'
params = {'location': 'VA'}
r1 = requests.get(url_type, headers=headers, params=params)
adopt_VA = r1.json()

#adopt_VA
for animal in adopt_VA['types']:
  print(animal['name'])
```

```python
#first additional enhancement, make a data frame that will extract all the animal types available for adoption in virginia.
# Extract the names of the available types of animals
animal_types = [animal['name'] for animal in adopt_VA['types']]
gender_types = [animal['genders'] for animal in adopt_VA['types']]
# Create a pandas dataframe with the animal types
df_adopt_VA = pd.DataFrame(gender_types, index = animal_types)

df_adopt_VA
```

```python
# What organizations can I adopt from in Maryland?
url_orga = 'https://api.petfinder.com/v2/organizations'
params = {'location': 'MD'}
r2 = requests.get(url_orga, headers=headers, params=params)
adopt_MD = r2.json()

for org_name in adopt_MD['organizations']:
  print(org_name['name'])
#adopt_MD
```

```python
#Second additional enhancement, make a data frame that will extract all the animal types available for adoption in virginia.
# Extract the names, cities, and ID of the organisation in maryland
names = [org_name['name'] for org_name in adopt_MD['organizations']]
ID = [org_id['id'] for org_id in adopt_MD['organizations']]
cities = [org_city['address']['city'] for org_city in adopt_MD['organizations']]
# Create a pandas dataframe with the name, ID and location of where we can adopt in maryland with cities as index.
df_adopt_MD = pd.DataFrame({'Names': names, 'ID': ID}, index = cities)

df_adopt_MD
```

```python
# Display the list of available dogs and breeds in a particular zip code (ask user for the zip code)
zip_code = input("\nEnter a zip code to search for available dogs: ")
url_zip = f'https://api.petfinder.com/v2/types/Dog/breeds'
params = {'location':f'{zip_code}', 'type': 'breeds'}
r4 = requests.get(url_zip, headers=headers, params=params)
available_dogs = r4.json()

print(f'\nDisplay list of available dog and breeds found for this zip code: {zip_code}\n')
for dog in available_dogs['breeds']:
    print(dog['name'])
```

```python
#Fourth additional enhancement, Display the list of Bird.
url_bird = 'https://api.petfinder.com/v2/types/Bird/breeds'
params = {'type': 'breeds'}
r5 = requests.get(url_bird, headers=headers, params=params)
bird = r5.json()

print(f'\nDisplay a list of available bird and breeds found\n')
for birds in bird['breeds']:
    print(birds['name'])
```
